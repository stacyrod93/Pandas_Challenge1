```python
import pandas as pd

purchase_csv = "Resources/purchase_data.csv"

purchase_data = pd.read_csv(purchase_csv)
purchase_data
```

```python
PlayerCount = len(purchase_data["SN"].unique())
playercount_disp = pd.DataFrame({"Player Count": [PlayerCount]})
playercount_disp
```

```python
Items = len(purchase_data["Item Name"].unique())
AvgPrice = purchase_data["Price"].mean()
AmountPurchase=len(purchase_data["Item Name"])
Revenue = purchase_data["Price"].sum()

Purchasing_Analysis_Total = pd.DataFrame({"Number of Unique Items": [Items],
                                           "Average Price": [AvgPrice],
                                           "Number of Purchases": [AmountPurchase],
                                           "Total Revenue": [Revenue]})

Purchasing_Analysis_Total["Average Price"] = Purchasing_Analysis_Total["Average Price"].map("${:.2f}".format)
Purchasing_Analysis_Total["Total Revenue"] = Purchasing_Analysis_Total["Total Revenue"].map("${:.2f}".format)
Purchasing_Analysis_Total = Purchasing_Analysis_Total[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

Purchasing_Analysis_Total
```

```python
dup_data = purchase_data.drop_duplicates(subset='SN', keep="first")
Gender = dup_data["Gender"].count()
Male = dup_data["Gender"].value_counts()['Male']
Female = dup_data["Gender"].value_counts()['Female']
NonDisclose = Gender - Male - Female

MalePer = (Male / Gender) * 100
FemalePer = (Female / Gender) * 100
NonPer = (NonDisclose / Gender) * 100

grouped_df = purchase_data.groupby(["Gender"])

purchCount = grouped_df["SN"].count()
purchPrice = grouped_df["Price"].mean()
purchValue = grouped_df["Price"].sum()

duplicate = purchase_data.drop_duplicates(subset='SN', keep="first")
grouped_dup = duplicate.groupby(["Gender"])

purchase = (grouped_df["Price"].sum() / grouped_dup["SN"].count())

Purchase_per_gender = pd.DataFrame({"Purchase Count": purchCount,
                              "Average Purchase Price": purchPrice,
                              "Total Purchase Price": purchValue,
                              "Average Purchase Price Per Person": purchase})


Purchase_per_gender["Average Purchase Price"] = Purchase_per_gender["Average Purchase Price"].map("${:.2f}".format)
Purchase_per_gender["Total Purchse Price"] = Purchase_per_gender["Total Purchase Price"].map("${:.2f}".format)
Purchase_per_gender["Average Purchase Price Per Person"] = Purchase_per_gender["Average Purchase Price Per Person"].map("${:.2f}".format)
Purchase_per_gender = Purchase_per_gender[["Purchase Count", "Average Purchase Price", "Total Purchase Price", "Average Purchase Price Per Person"]]
Purchase_per_gender
```

```python
bins = [0,10,15,20,25,30,35,40,200]
binLab = ['< 10', '10 - 14', '15 - 19', '20 - 24', '25 - 29', '30 - 34', '35 - 39', '40+']

binner_df = purchase_data.copy()
binner_df["Age Groups"] = pd.cut(binner_df["Age"], bins, labels=binLab)
group_bin = binner_df.groupby(["Age Groups"])

binnerCount = group_bin["SN"].count()
countTotal = purchase_data["SN"].count()
percentage = (binnerCount / countTotal) * 100
percentage

Age_percent = pd.DataFrame({"Total Count": binnerCount,
                         "Percentage of Players": percentage})

Age_percent["Percentage of Players"] = Age_percent["Percentage of Players"].map("{:.2f}%".format)
Age_percent.head(10)
```

```python
groupedBySN = purchase_data.groupby(["SN"])
groupedCount = groupedBySN["Item ID"].count()
groupedTotal = groupedBySN["Price"].sum()
groupedAvg = (groupedTotal / groupedCount)


Demo = pd.DataFrame({"Purchase Count": groupedCount,
                         "Average Purchase Price": groupedAvg,
                         "Total Purchase Value": groupedTotal})

Demo = Demo.sort_values("Total Purchase Value", ascending=False) 
Demo["Average Purchase Price"] = Demo["Average Purchase Price"].map("${:.2f}".format)
Demo["Total Purchase Value"] = Demo["Total Purchase Value"].map("${:.2f}".format)
Demo = Demo[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
Demo.head()
```

```python
groupItem = purchase_data.groupby(["Item ID", "Item Name"])
groupItemCount = groupItem["SN"].count()
groupPriceSum = groupItem["Price"].sum()
groupItemPrice = (groupPriceSum / groupItemCount)
groupItemValue = (groupItemPrice * groupItemCount)

Popular= pd.DataFrame({"Purchase Count": groupItemCount,
                          "Item Price": groupItemPrice,
                          "Total Purchase Value": groupItemValue})

Popular = Popular.sort_values("Purchase Count", ascending=False) 
Popular["Item Price"] = Popular["Item Price"].map("${:.2f}".format)
Popular["Total Purchase Value"] = Popular["Total Purchase Value"].map("${:.2f}".format)
Popular = Popular[["Purchase Count", "Item Price", "Total Purchase Value"]]
Popular.head()
```
